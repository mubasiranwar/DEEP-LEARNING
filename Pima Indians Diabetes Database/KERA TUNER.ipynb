```python
X=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Dropout
from keras import Sequential
```

```python
model=Sequential()
model.add(Dense(32,activation='relu',input_dim=8))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train.to_numpy(),batch_size=32,epochs=100,validation_data=(X_test,y_test.to_numpy()))
```

```python
import keras_tuner as kt
```

```python
def model_building(hp):
  model=Sequential()
  model.add(Dense(32,activation='relu',input_dim=8))
  model.add(Dense(1,activation='sigmoid'))

  optimizer=hp.Choice('optimizer',values=['adam','sgd','rmsprop'])
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tunr=kt.RandomSearch(model_building,objective='val_accuracy',max_trials=5)
```

```python
tunr.search(X_train,y_train.to_numpy(),epochs=5,validation_data=(X_test,y_test.to_numpy()))
```

```python
tunr.get_best_hyperparameters()[0].values
```

```python
model=tunr.get_best_models(num_models=1)[0]
```

```python
model.summary()
```

```python
def model_bulid(hp):
  model=Sequential()
  units=hp.Int('units',min_value=8,max_value=64,step=8)
  model.add(Dense(units=units,activation='relu',input_dim=8))
  model.add(Dense(1,activation='sigmoid'))

  optimizer=hp.Choice('optimizer',values=['adam','sgd','rmsprop'])
  model.compile(optimizer=optimizer,loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```python
tuner=kt.RandomSearch(model_bulid,objective='val_accuracy',max_trials=5)
tuner.search(X_train,y_train.to_numpy(),epochs=5,validation_data=(X_test,y_test.to_numpy()))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model=tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(X_train,y_train.to_numpy(),initial_epoch=5,epochs=100,validation_data=(X_test,y_test.to_numpy()))
```
